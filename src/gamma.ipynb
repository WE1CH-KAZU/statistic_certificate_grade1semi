```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

```

* a=1 : 魚が1匹釣れるまでの時間（指数分布）
* a=2 : 魚が2匹釣れるまでの累計時間
* a=3 : 魚が3匹釣れるまでの累計時間
  
* b=1 (λ=1/b) : 魚が1匹釣れるまでの平均待ち時間（1/1 = 1hrs(又は1min. 単位は問題の設定による)
* b=2 (λ=1/b) : 魚が1匹釣れるまでの平均待ち時間（1/2 = 0.5hrs）
* b=3 (λ=1/b) : 魚が1匹釣れるまでの平均待ち時間（1/3 = 0.333…hrs）

```python
# x 軸の範囲
x = np.linspace(0, 10, 1000)

# ガンマ分布のパラメータ (a, b) の組み合わせ
params_1 = [(1, 1), (1, 2), (1, 3)]
params_2 = [(2, 1), (2, 2), (2, 3)]
params_3 = [(3, 1), (3, 2), (3, 3)]

# グラフの作成
plt.figure(figsize=(8, 6))

for a, b in params_1:
    # ガンマ分布の確率密度関数 (PDF)
    y = gamma.pdf(x, a, scale=1/b)
    plt.plot(x, y, label=f'a={a}, b={b}')

plt.title('Gamma Distribution for Different a and b')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

* bが大きくなる（次釣れるまでの時間が短くなる）につれて、時間xが小さい時（早い段階）に確率密度が高くなっている

```python
# グラフの作成
plt.figure(figsize=(8, 6))

for a, b in params_2:
    # ガンマ分布の確率密度関数 (PDF)
    y = gamma.pdf(x, a, scale=1/b)
    plt.plot(x, y, label=f'a={a}, b={b}')

plt.title('Gamma Distribution for Different a and b')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

* 2匹釣れるまでの累計時間になると、時間xがある程度過ぎたころに確率密度の高い所が現れる

```python
# グラフの作成
plt.figure(figsize=(8, 6))

for a, b in params_3:
    # ガンマ分布の確率密度関数 (PDF)
    y = gamma.pdf(x, a, scale=1/b)
    plt.plot(x, y, label=f'a={a}, b={b}')

plt.title('Gamma Distribution for Different a and b')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```
