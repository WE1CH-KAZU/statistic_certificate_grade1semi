```python
import numpy as np
from scipy.stats import kurtosis
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# データ生成
np.random.seed(0)
size = 10000
```

```python
# 分布データ
normal = np.random.normal(loc=0, scale=1, size=size)
leptokurtic = np.random.laplace(loc=0, scale=1/np.sqrt(2), size=size)
platykurtic = np.random.uniform(low=-np.sqrt(3), high=np.sqrt(3), size=size)

# 尖度（通常の定義では正規分布で3となるので補正なしで計算）
kurt_norm = kurtosis(normal, fisher=False)
kurt_lepto = kurtosis(leptokurtic, fisher=False)
kurt_platy = kurtosis(platykurtic, fisher=False)
```

```python
# 尖度の計算（Fisher補正あり: 正規分布=0 となる）
print("Normal kurtosis:", kurtosis(normal))           # ≈ 0
print("Leptokurtic kurtosis:", kurtosis(leptokurtic)) # > 0
print("Platykurtic kurtosis:", kurtosis(platykurtic)) # < 0
```

```python
# 線グラフ（KDE;カーネル密度推定）で比較
sns.kdeplot(normal, label=f'Normal = {kurt_norm:.2f}', linewidth=2)
sns.kdeplot(leptokurtic, label=f'Leptokurtic = {kurt_lepto:.2f}', linewidth=2)
sns.kdeplot(platykurtic, label=f'Platykurtic = {kurt_platy:.2f}', linewidth=2)

plt.title("Comparison of Distributions and Kurtosis")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
exponentialkurtosis = np.random.exponential(scale=1, size=size) # 指数分布、期待値1
Tdistkurtosis = np.random.standard_t(df=5, size=size) # T分布 自由度5
Beta1kurtosis = np.random.beta(a=0.5, b=0.5, size=size) # beta分布(a,b)=(0.5,0.5)
Beta2kurtosis = np.random.beta(a=2, b=5, size=size) # beta分布(a,b)=(2,5)
Beta3kurtosis = np.random.beta(a=5, b=2, size=size) # beta分布(a,b)=(5,2)

# 尖度（通常の定義では正規分布で3となるので補正なしで計算）
kurt_exp = kurtosis(exponentialkurtosis, fisher=False)
kurt_tdist = kurtosis(Tdistkurtosis, fisher=False)
kurt_beta1 = kurtosis(Beta1kurtosis, fisher=False)
kurt_beta2 = kurtosis(Beta2kurtosis, fisher=False)
kurt_beta3 = kurtosis(Beta3kurtosis, fisher=False)
```

```python
# 尖度の計算（Fisher補正あり: 正規分布=0 となる）
print("Exponential kurtosis:", kurtosis(exponentialkurtosis)) # > 0
print("T-distribution kurtosis:", kurtosis(Tdistkurtosis)) # >> 0
print("Beta1 kurtosis:", kurtosis(Beta1kurtosis)) # < 0
print("Beta2 kurtosis:", kurtosis(Beta2kurtosis)) # < 0
print("Beta3 kurtosis:", kurtosis(Beta3kurtosis)) # < 0
```

```python
# 線グラフ（KDE;カーネル密度推定）で比較
sns.kdeplot(exponentialkurtosis, label=f'Exponential = {kurt_exp:.2f}', linewidth=2)
sns.kdeplot(Tdistkurtosis, label=f'T-distribution = {kurt_tdist:.2f}', linewidth=2)
sns.kdeplot(normal, label=f'Normal = {kurt_norm:.2f}', linewidth=2)

plt.title("Comparison of Distributions and Kurtosis")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.kdeplot(Beta1kurtosis, label=f'Beta1(0.5,0.5) = {kurt_beta1:.2f}', linewidth=2)
sns.kdeplot(Beta2kurtosis, label=f'Beta2(2,5) = {kurt_beta2:.2f}', linewidth=2)
sns.kdeplot(Beta3kurtosis, label=f'Beta3(5,2) = {kurt_beta3:.2f}', linewidth=2)
sns.kdeplot(normal, label=f'Normal = {kurt_norm:.2f}', linewidth=2)

plt.title("Comparison of Distributions and Kurtosis")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
